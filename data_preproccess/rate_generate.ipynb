```python
import pandas as pd
import numpy as np
```

# Generate RN & RD 

```python
path = '/home/users/YongsungKwon/workplace/Yongpyter/Tuberculosis_hospital_optimization/data_result/raw_data/'
D_sido = pd.read_csv(path+'2014_2022_sido_10age_fatalities.txt',sep='\t')
N_sido = pd.read_csv(path+'2014_2022_sido_10age_patients.txt',sep='\t')
```

```python
for y in range(2014, 2023):
    year=str(y) #year = '2014'
    data_sigungu = pd.read_csv(path+year+'.txt',sep='\t')
    D_sido_year = D_sido.loc[:, ["sido", "age", year]]
    N_sido_year = N_sido.loc[:, ["sido", "age", year]]
    D_sido_year['D_sum'] = D_sido_year.groupby('sido')[year].transform('sum')
    D_sido_year['RD'] = D_sido_year[year]/D_sido_year['D_sum']
    N_sido_year['N_sum'] = N_sido_year.groupby('sido')[year].transform('sum')
    N_sido_year['RN'] = N_sido_year[year]/N_sido_year['N_sum']

    D_pivot = D_sido_year.pivot(index="sido", columns="age", values="RD")
    D_pivot.columns = ["RD" + str(col) for col in D_pivot.columns]
    D_pivot.reset_index(inplace=True)
    N_pivot = N_sido_year.pivot(index="sido", columns="age", values="RN")
    N_pivot.columns = ["RN" + str(col) for col in N_pivot.columns]
    N_pivot.reset_index(inplace=True)

    new_data = pd.DataFrame()
    new_data['sido'] = data_sigungu['#sido']
    new_data['sigungu'] = data_sigungu['sigungu']
    new_data['h'] = data_sigungu['h']
    new_data['A'] = data_sigungu['A']
    new_data['N'] = data_sigungu['N']
    new_data['D'] = data_sigungu['D']
    
    new_data = new_data.merge(N_pivot, on="sido", how="left")
    new_data = new_data.merge(D_pivot, on="sido", how="left")
    new_data.to_csv(path+year+'_RNRD.txt',sep='\t', index=False)
```
