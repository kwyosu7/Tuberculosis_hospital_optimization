```python
P_grouped = P_data.groupby("#sido", as_index=False)[["0_9", "10_19", "20_29", "30_39", "40_49", "50_59", "60_69", "70_79", "80_"]].sum()
```

```python
age = P_data.columns.tolist()[6:]
for l,e in P_grouped.iterrows():
    P_local = e.loc[age].tolist()
    plt.plot(age, P_local, label = e['#sido'])

plt.legend()
plt.yscale('log')
plt.show()
```
